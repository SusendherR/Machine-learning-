```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset =pd.read_csv('Mall_Customers.csv')
dataset
```

```python
x = dataset.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import AffinityPropagation
list1=[]
for i in range(1,11):
    cluster=AffinityPropagation(damping=0.5)
    cluster.fit(x)
    list1.append(cluster.labels_)
plt.title('AffinityPropagation')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```python
import lightgbm as LGBMRegressor
import numba as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
depandent=dataset[['charges']]
```

```python
independent
```

```python
depandent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,depandent,test_size=0.30,random_state=0)
```

```python
from lightgbm import LGBMRegressor
regressor=LGBMRegressor(n_estimators=50,boosting_type='gbdt',)
regressor=regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_LGBMRegressor.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
load_model=pickle.load(open("finalized_LGBMRegressor.sav",'rb'))
result=load_model.predict([[28,33.000,3,1,0]])
```
